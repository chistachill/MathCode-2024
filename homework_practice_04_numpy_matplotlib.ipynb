```python
#2.1
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 250)
y = x**2 * np.sin(1/x**2) + x

plt.plot(x, y, color='pink')
plt.xlabel('x')
plt.ylabel('y')
plt.title('График функции x^2*sin(1/x^2) + x')
plt.legend('y = x^2 * sin(1/x^2) + x')
plt.grid(True)
plt.show()
```

```python
#2.3.1
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import airy

x = np.linspace(1, 20, 200)
ai, aip, bi, bip = airy(x)

plt.plot(x, ai, label='Ai(x)')
plt.plot(x, bi, label='Bi(x)')
plt.plot(x, aip, label="Ai'(x)")
plt.plot(x, bip, label="Bi'(x)")
plt.xlabel('x')
plt.ylabel('y - значения функции ')
plt.title('airy - Воздушные функции и их производные. ')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#2.3.2
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import elliprc

x_values = np.linspace(0, 2, 100)
y_values = [0.1, 0.3, 0.5, 0.7, 0.9, 1.1, 1.3, 1.5, 1.7, 1.9]
plt.figure()

for y in y_values:
    elliprc_values = [elliprc(x, y) for x in x_values]
    plt.plot(x_values, elliprc_values, label=f'y={y}')

plt.legend()
plt.title('График elliprc(x, y) для различных значений y')
plt.xlabel('x')
plt.ylabel('elliprc(x, y)')
plt.grid(True)
plt.show()
```

```python
#2.3.3
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import itj0y0

x_values = np.linspace(0, 20, 100)
itj0_values = itj0y0(x_values)[0]
y0_values = itj0y0(x_values)[1]

plt.figure(figsize=(10, 6))
plt.plot(x_values, itj0_values, label='Модифицированная функция Бесселя J0', color='pink')
plt.plot(x_values, y0_values, label='Функция Бесселя Y0', color='lightblue')
plt.xlabel('x')
plt.ylabel('Значения')
plt.title('Модифицированные функции и функции Бесселя')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
def f(y):
    return 1/(y**2-2*y-8)

def rectangle_method(f, a, b, N):
    delta_x = (b - a) / N
    x_k = np.linspace(a, b, N+1)
    integral = delta_x * np.sum(f(x_k[:-1]))
    return integral

def trapezoid_method(f, a, b, N):
    delta_x = (b - a) / N
    x_k = np.linspace(a, b, N+1)
    integral = (delta_x / 2) * (f(a) + 2*np.sum(f(x_k[1:-1])) + f(b))
    return integral

def simpson_method(f, a, b, N):
    delta_x = (b - a) / N
    x_k = np.linspace(a, b, N+1)
    integral = (delta_x / 3) * (f(a) + 2*np.sum(f(x_k[2:N:2])) + 4*np.sum(f(x_k[1:N:2]) + f(b)))
    return integral

a = 2
b = 3
N = 10

integral_rectangle = rectangle_method(f, a, b, N)
integral_trapezoid = trapezoid_method(f, a, b, N)
integral_simpson = simpson_method(f, a, b, N)

integral_rectangle, integral_trapezoid, integral_simpson
```
